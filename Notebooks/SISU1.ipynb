```python
#Importing the libraries needed for the project
import pandas as pd

#Reading the CSV file and getting the dataset
sisu1 = pd.read_csv("../Data/ListagemChamadaRegular_2020-1-Filtrado.csv", sep=';')

sisu1['NO_CURSO'].unique()
```

```python
#Getting the data about the courses at the Federal University of Espírito Santo Technological Center 
#Computer Science
#Environmental Engineering
#Civil Engineering
#Computer Engineering
#Production Engineering
#Electrical Engineering
#Mechanical Engineering

techcenter = sisu1.loc[(sisu1['NO_CURSO'] == "ENGENHARIA CIVIL") | (sisu1['NO_CURSO'] == "ENGENHARIA AMBIENTAL") | (sisu1['NO_CURSO'] == "CIÊNCIA DA COMPUTAÇÃO") | (sisu1['NO_CURSO'] == "ENGENHARIA DA COMPUTAÇÃO") | (sisu1['NO_CURSO'] == "ENGENHARIA DE PRODUÇÃO") | (sisu1['NO_CURSO'] == "ENGENHARIA ELÉTRICA") | (sisu1['NO_CURSO'] == "ENGENHARIA MECÂNICA")]

techcenter.shape
```

```python
#Getting the data of the names of the studentes subscribed in SISU in any course at the Federal University of Espírito Santo Technological Center
names = techcenter['NO_INSCRITO']

#Printing the names in a text file
name_files = open('../Files/SISU_1/NomeInscritos.txt', 'w')

for name in names:
    name_files.write(f"{name}\n")

name_files.close()
```

```python
#Getting the data of the courses at the Federal University of Espírito Santo Technological Center
courses = techcenter['NO_CURSO']

#Printing the number of subscribed students in each course in a text file
course_files = open('../Files/SISU_1/CursoInscritos.txt', 'w')

civil = 0
computer = 0
ele = 0
env = 0
sci = 0
prod = 0
mec = 0

for course in courses:
    if course == "ENGENHARIA CIVIL":
        civil = civil + 1
    elif course == "ENGENHARIA DA COMPUTAÇÃO":
        computer = computer + 1
    elif course == "ENGENHARIA ELÉTRICA":
        ele = ele + 1
    elif course == "ENGENHARIA AMBIENTAL":
        env = env + 1
    elif course == "CIÊNCIAS DA COMPUTAÇÃO":
        sci = sci + 1
    elif course == "ENGENHARIA DE PRODUÇÃO":
        prod = prod + 1
    else:
        mec = mec + 1

course_files.write(f"ENGENHARIA CIVIL: {civil}\n")
course_files.write(f"ENGENHARIA AMBIENTAL: {env}\n")
course_files.write(f"ENGENHARIA DA COMPUTAÇÃO: {computer}\n")
course_files.write(f"ENGENHARIA DE PRODUÇÃO: {prod}\n")
course_files.write(f"ENGENHARIA ELÉTRICA: {ele}\n")
course_files.write(f"ENGENHARIA MECÂNICA: {mec}\n")
course_files.write(f"CIÊNCIA DA COMPUTAÇÃO: {sci}\n")

course_files.close()
```

```python
#Printing the general info of the students subscribed in SISU
geral = open('../Files/SISU_1/ResumoGeral.txt', 'w')

for i, info in techcenter.iterrows():
    geral.write(f"{info['NO_INSCRITO']} ({info['NO_CURSO']}) - ")

    if info['ST_APROVADO'] == 'S':
        geral.write("Aprovado(a) 2020/1 - ")

        if info['ST_MATRICULA'] == "PENDENTE":
            geral.write("Matrícula Pendente\n")
        elif info['ST_MATRICULA'] == "EFETIVADA":
            geral.write("Matrícula Efetivada\n")
        else:
            geral.write("Candidato não compareceu para a matrícula\n")
        
    else:
        geral.write("Não Aprovado(a)\n")

geral.close()
```
